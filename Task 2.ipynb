```python
# 2.Create histograms and boxplots for numeric features.
```

```python
# selecting the only numerical features
```

```python
numeric_cols=df.select_dtypes(include='number').columns
```

```python
numeric_cols
```

```python
# creatings histograms for numeric columns
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# creating box plots of numeric columns
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
# 4.Identify patterns, trends, or anomalies in the data.
```

```python
# identifying patterns by grouping upon mean
```

```python
df = sns.load_dataset("titanic")
df
```

```python
#  Survival Rate by passenger class(pclass) Group
```

```python
print(df.groupby('pclass')['survived'].mean())
```

```python
sns.barplot(x='pclass', y='survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# Survival Rate by Age Group
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 12, 18, 30, 60, 80], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
print(df.groupby('age_group')['survived'].mean())
```

```python
sns.barplot(x='age_group', y='survived', data=df)
```

```python
# detecting anomalies
```

```python
print(df[df['fare'] > 300])
```

```python
sns.boxplot(x='pclass', y='fare', data=df)
```

```python
# 5.Make basic feature-level inferences from visuals
```

```python
# Fare vs. Survival (Box Plot)
```

```python
sns.boxplot(x='survived', y='fare', data=df)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
# Survival by Class and Gender (Heatmap)
```

```python
pivot=df.pivot_table(index='pclass',columns='sex',values='survived')
sns.heatmap(pivot,annot=True,cmap='Blues')
plt.title("Survival Rate by Class and Gender")
plt.show()
```
